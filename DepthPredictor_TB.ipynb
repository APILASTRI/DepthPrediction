```python
def load_csv(path):
    """ 
    returns a dict from a CSV of format:
        /a/abbey 0
        /a/airport_terminal 1
        /a/alley 2
        etc
    """
    f = open(path)
    csvf = csv.reader(f)
    labels = {}
    for row in csvf:
        tmp = row[0].split(' ',1)
        labels[int(tmp[1])] = tmp[0]
    f.close()
    
    return labels
```
