```python
# Before using:
# -bash $CAFFE_HOME/tools/extra/parse_log.sh /tmp/caffe.INFO
# (creates two tables, caffe.INFO.test, caffe.INFO.train)

train = {}
tmp = np.loadtxt('caffe.INFO.train')
train['iters'], train['seconds'], train['loss'], train['learn_rate'] = tmp[:,0], tmp[:,1], tmp[:,2], tmp[:,3]
# Iters Seconds TrainingLoss LearningRate

plt.subplot(211)
plt.plot(train['iters'],train['loss'],'b')
plt.ylabel('loss')
plt.title('Train Set')
plt.ylim(0,15)

plt.subplot(212)
plt.plot(train['iters'],train['learn_rate'],'b')
plt.xlabel('iters')
plt.ylabel('learn rate')
plt.show()

test = {}
tmp = np.loadtxt('caffe.INFO.test')
test['iters'], test['seconds'], test['accuracy'], test['loss'] = tmp[:,0], tmp[:,1], tmp[:,2], tmp[:,3]
# Iters Seconds TrainingLoss LearningRate
plt.plot(test['iters'],test['accuracy'],'b')
plt.title('Validation Set')
plt.xlabel('iters')
plt.ylabel('learn rate')
plt.show()
```

```python
# Plot histograms of depths in a dataset:

in_file = 'train/NYUv2/train_full_167_v2/index_16_norm.txt'


import csv

f = open(in_file,'r')
depths = []
csvf = csv.reader(f)

for row in csvf:
    tmp = row[0].split(' ',1)
    depths.append(tmp[1])
print np.unique(depths)
plt.plot(np.bincount(depths)/(0.01*len(depths)))
plt.ylabel('Occurance Percent')
plt.xlabel('Bin Number')
plt.xlim([1,len(np.unique(depths))])
```
